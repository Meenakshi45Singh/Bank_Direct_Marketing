# Categorical Features

```python
catFeatures=df.select_dtypes(include=['object']).columns
print("Total no of Categorical Features = ",catFeatures.shape[0])
print("Name of Categorical Features = ",catFeatures)
```

```python
catDf=df[catFeatures]
catDf.head(2)
```

```python
#Print Unique Values for each columns
for col in catDf.columns:
    print(f"{col}: {catDf[col].unique()}\n")
```

```python
#Encoding Categorical Variables
catDf['Job']=catDf['Job'].map({'management':0, 'technician':1, 'entrepreneur':2, 'blue-collar':3, 'unknown':4,
         'retired':5, 'admin':6, 'services':7, 'self-employed':8, 'unemployed':9, 'housemaid':10,
         'student':11})

catDf['Marital']=catDf['Marital'].map({'married':0, 'single':1, 'divorced':2})

catDf['Education']=catDf['Education'].map({'tertiary':0, 'secondary':1, 'unknown':2, 'primary':3})

catDf['Default']=catDf['Default'].map({'no':0, 'yes':1})

catDf['Housing']=catDf['Housing'].map({'no':0, 'yes':1})

catDf['Loan']=catDf['Loan'].map({'no':0, 'yes':1})

catDf['Contact']=catDf['Contact'].map({'unknown':0, 'cellular':1, 'telephone':2})

catDf['Month']=catDf['Month'].map({'jan':0, 'feb':1, 'mar':2, 'apr':3, 'may':4, 'jun':5, 'jul':6, 'aug':7,
                                   'sep':8, 'oct':9, 'nov':10, 'dec':11})

catDf['Poutcome']=catDf['Poutcome'].map({'unknown':0, 'failure':1, 'other':2, 'success':3})

catDf['Y']=catDf['Y'].map({'no':0, 'yes':1})
```

```python
catDf.shape
```

```python
catDf.head()
```

```python
newDf = pd.concat([catDf, numDf], axis = 1)
newDf.head()
```

```python
X=newDf.drop('Y',axis=1)
y=newDf['Y']
```

# Model Building

# Random Forest Classifier

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0, test_size=0.2)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rdc = RandomForestClassifier()
rdc.fit(X_train, y_train)
y_pred_rdc = rdc.predict(X_test)
```

```python
accrdc = accuracy_score(y_test, y_pred_rdc)
conrdc = confusion_matrix(y_test, y_pred_rdc)
reportrdc = classification_report(y_test, y_pred_rdc)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
print(f"Accuracy Score of Random Forest is : {accrdc}")
print("-------------------------------------------------")
print(f"Confusion Matrix : \n{conrdc}")
print("-------------------------------------------------")
print(f"Classification Report : \n{reportrdc}")
```
